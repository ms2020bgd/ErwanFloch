```python
def find_contributors():
    L = []
    i = 0
    url = "https://gist.github.com/paulmillr/2657075"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for e in soup.find_all("td"):
        b = e.parent
        c = b.find("td").text
        if (c not in L) and (i<256):
            L.append(c)
            i = i + 1
    return L

liste_contributors=find_contributors()
print(find_contributors())
print()
print("Taille de liste : {}".format(len(liste_contributors)))
```

```python
# Test pour le user1
contributors_list = find_contributors()
#user1 = liste_contributors[0].split(" ")[0]
users_dict = {} # Dictionary for user i: the key is the username; the values are the list of the number of stars 

def find_stars():

    for i in range(0,25):
        user = contributors_list[i].split(" ")[0]
        url = "https://api.github.com/users/" + user + "/repos"          # user n°i
        response = requests.get(url)
        data = json.loads(response.text)                                                                
    
        L=[]    # List containing the number of stars
        # retrieve the number of stars for the repository r
        for r in range(0, len(data)-1):
            L.append(data[r]['stargazers_count'])                                                   

        users_dict[user] = L   
    return users_dict

users_dict = find_stars()
```

**Computing the averages**

```python
users_dict_avg = {}
for k in users_dict.keys():
    avg = 0
    for v in users_dict[k]:
        avg = avg + v
    users_dict_avg[k] = avg / len (users_dict[k])

users_dict_avg
```
