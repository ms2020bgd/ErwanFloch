```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://fr.wikipedia.org/wiki/Liste_des_communes_de_France_les_plus_peuplées'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
def liste_des_communes():
a = soup.find('table')                     # 'table' = tableau

ligne_tableau = a.find_all('tr')          # 'tr' = ligne de tableau

Liste_communes = []

for ligne in ligne_tableau[1:]:    # on exclut la première ligne
    cellule = ligne.find_all('td') # 'td' = cellule de tableau
    ville = cellule[1].find('a')
    Liste_communes.append(ville.text)
    return Liste_communes

print()
```
